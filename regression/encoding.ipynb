## Target Encoder

```python
# encoder=TargetEncoder()

# encoder=TargetEncoder(cols=['Employment_Status'])
'''
when we have mixed column names(numeric and categorical)  we use this "cols"1 parameter. 
'''

encoder=TargetEncoder(cols=['Employment_Status'],handle_unknown='value')
'''
Default handle_unknown = 'error'
handle_unknown = 'value' --> Here,"value" works taking average of all the target column and assign it to unknown labels or values. 
'''

```
